The sequence of shocks {εt} is assumed to be IID and standard normal.)
Set T=200 and α/A=0.9.

```python
T=200
A=0.9
x=np.empty(T+1)
x[0]=0
for t in range(T):
    x[t+1]=A*x[t]+np.random.randn()
plt.plot(x)

```

# Exercise 2
Starting with your solution to exercise 2, plot three simulated time series, one for each of the cases α=0, α=0.8 and α=0.98.

Use a for loop to step through the α values.

If you can, add a legend, to help distinguish between the three time series.

```python
α_values = [0.0, 0.8, 0.98]
T = 200
x = np.empty(T+1)

for α in α_values:
    x[0] = 0
    for t in range(T):
        x[t+1] = α * x[t] + np.random.randn() # if x is |x| use np.abs(x[t])
    plt.plot(x, label=f'$\\alpha = {α}$')

plt.legend()
plt.show()
```

```python
def generate_value(n):
    e_values=[]
    for i in range(n):
        e=np.random.randn()
        e_values.append(e)
    return e_values
```

# Exercise 1
Here is some code that needs improving.

It involves a basic supply and demand problem.

Supply is given by

qs(p)=exp(αp)−β.
The demand curve is

qd(p)=γp−δ.
The values α, β, γ and δ are parameters

The equilibrium p∗ is the price such that qd(p)=qs(p).

We can solve for this equilibrium using a root finding algorithm. Specifically, we will find the p such that h(p)=0, where

h(p):=qd(p)−qs(p)
This yields the equilibrium price p∗. From this we get the equilibrium price by q∗=qs(p∗)
The parameter values will be

α=0.1
β=1
γ=1
δ=1

```python
from scipy.optimize import brentq
```

```python
def f(p):
    return p**(-1) - (np.exp(0.1 * p) - 1)  # demand - supply
p_star = brentq(f, 2, 4)
q_star = np.exp(0.1 * p_star) - 1

print('Equilibrium price is',p_star)
print('Equilibrium quantity is',q_star)
```

```python
# Plot
grid = np.linspace(2,4,100 )
fig, ax = plt.subplots()

qs = np.exp(0.1 * grid) - 1
qd = grid**(-1)


ax.plot(grid, qd, 'b-', lw=2, label='demand')
ax.plot(grid, qs, 'g-', lw=2, label='supply')

ax.set_xlabel('price')
ax.set_ylabel('quantity')
ax.legend(loc='upper center')

plt.show()
```
